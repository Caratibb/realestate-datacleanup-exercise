#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = ds[ds['level5'].isin(populations_of_interest)]
plt.figure(figsize=(12, 8))
for population in populations_of_interest:
    subset_population = south_belt_df[south_belt_df['level5'] == population]
    normalized_prices = (subset_population['price'] - subset_population['price'].mean()) / subset_population['price'].std()
    plt.hist(normalized_prices, bins=20, alpha=0.5, label=population)
plt.title('Normalized Prices Histograms for South Belt of Madrid')
plt.xlabel('Normalized Prices')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
populations_of_interest = ["Getafe", "Alcorcón"]
south_belt_df = ds[ds['level5'].isin(populations_of_interest)]
south_belt_df['price_per_m2'] = south_belt_df['price'] / south_belt_df['surface']
mean_ppm_getafe = south_belt_df[south_belt_df['level5'] == 'Getafe']['price_per_m2'].mean()
mean_ppm_alcorcon = south_belt_df[south_belt_df['level5'] == 'Alcorcón']['price_per_m2'].mean()
print("Mean Price per Square Meter (price/m2) in Getafe: {mean_ppm_getafe:.2f} euros")
print("Mean Price per Square Meter (price/m2) in Alcorcón: {mean_ppm_alcorcon:.2f} euros")
if mean_ppm_getafe > mean_ppm_alcorcon:
    print("Getafe tends to have a higher average price per square meter than Alcorcón.")
elif mean_ppm_getafe < mean_ppm_alcorcon:
    print("Alcorcón tends to have a higher average price per square meter than Getafe.")
else:
    print("The average price per square meter is the same in both Getafe and Alcorcón.")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import matplotlib.pyplot as plt
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = ds[ds['level5'].isin(populations_of_interest)]
south_belt_df['price_per_m2'] = south_belt_df['price'] / south_belt_df['surface']
plt.figure(figsize=(10, 6))
for population in populations_of_interest:
    subset_population = south_belt_df[south_belt_df['level5'] == population]
    plt.bar(population, subset_population['price_per_m2'].mean(), label=population)
plt.title('Average Price per Square Meter in Different Populations')
plt.xlabel('Population')
plt.ylabel('Average Price per Square Meter (price/m2)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_df = ds[ds['level5'].isin(populations_of_interest)]
colors = {'Fuenlabrada': 'blue', 'Leganés': 'green', 'Getafe': 'red', 'Alcorcón': 'purple'}
plt.figure(figsize=(10, 8))
for population in populations_of_interest:
    subset_population = south_belt_df[south_belt_df['level5'] == population]
    plt.scatter(subset_population['longitude'], subset_population['latitude'], color=colors[population], label=population, alpha=0.7)
plt.title('Coordinates of South Belt of Madrid by Population')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
